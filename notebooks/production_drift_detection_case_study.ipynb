## 1. Initialization

Let's set up the Production Drift Detection detectors, monitors, and generators.

```python
from production_drift_detection.data.synthetic_drift import DriftGenerator
from production_drift_detection.detectors.kl import KLDivergenceDetector
from production_drift_detection.detectors.psi import PSIDetector
from production_drift_detection.detectors.mmd import MMDDetector
from production_drift_detection.detectors.adwin import ADWINDetector
from production_drift_detection.monitors.stream_monitor import StreamMonitor
from production_drift_detection.monitors.confidence_monitor import ConfidenceMonitor
from production_drift_detection.correlation.confidence_drift import ConfidenceDriftCorrelation
from production_drift_detection.evaluation.metrics import (
    evaluate_detector, 
    compute_detection_latency,
    compute_false_positive_rate,
)

print("All imports successful")
```

```python
# Configuration
N_FEATURES = 5
N_REFERENCE = 2000
N_BATCHES = 30
BATCH_SIZE = 100
DRIFT_START_BATCH = 10
RANDOM_STATE = 42

print(f"Configuration:")
print(f"  Features: {N_FEATURES}")
print(f"  Reference samples: {N_REFERENCE}")
print(f"  Batches: {N_BATCHES}")
print(f"  Batch size: {BATCH_SIZE}")
print(f"  Drift starts at batch: {DRIFT_START_BATCH}")
```

## 2. Generate Reference Data and Create Detectors

```python
# Generate reference distribution
generator = DriftGenerator(
    n_features=N_FEATURES,
    n_reference=N_REFERENCE,
    random_state=RANDOM_STATE,
)
reference_data = generator.generate_reference()

print(f"Reference data shape: {reference_data.shape}")
print(f"Reference mean: {np.mean(reference_data, axis=0)}")
print(f"Reference std: {np.std(reference_data, axis=0)}")
```

```python
# Create detectors
detectors = {
    'kl': KLDivergenceDetector(threshold=0.1, is_categorical=False, n_bins=20),
    'psi': PSIDetector(threshold=0.1, n_bins=10, per_feature=True),
    'mmd': MMDDetector(threshold=0.05, subsample=200),
    'adwin': ADWINDetector(threshold=0.1, delta=0.05),
}

# Fit detectors
for name, detector in detectors.items():
    detector.fit(reference_data)
    print(f"{name.upper()}: Fitted — {detector.summary()['current_status']}")
```

## 3. Simulate Production with Drift Injection

We'll generate production batches with increasing drift magnitude.

```python
rng = np.random.default_rng(RANDOM_STATE)
batches = []
drift_magnitude = 0.0

for batch_idx in range(N_BATCHES):
    if batch_idx < DRIFT_START_BATCH:
        # Clean data
        batch = np.zeros((BATCH_SIZE, N_FEATURES))
        for col in range(N_FEATURES):
            mean = rng.uniform(-1, 1)
            std = rng.uniform(0.5, 1.0)
            batch[:, col] = rng.normal(mean, std, BATCH_SIZE)
    else:
        # Increasing drift
        drift_magnitude += 0.12
        batch = generator.covariate_shift(
            n_samples=BATCH_SIZE,
            shift_magnitude=drift_magnitude,
        )
    batches.append(batch)

print(f"Generated {len(batches)} batches")
print(f"Drift magnitude ranges from 0 to {drift_magnitude:.2f}")
```

## 4. Run Monitoring Pipeline

Process all batches through StreamMonitor and ConfidenceMonitor.

```python
# Initialize monitors
stream_monitor = StreamMonitor(detectors=detectors)
confidence_monitor = ConfidenceMonitor()
correlation = ConfidenceDriftCorrelation(max_lag=5)

# Fit stream monitor
stream_monitor.fit(reference_data)
print("Monitors initialized and fitted.")
```

```python
# Process batches
all_results = []
for batch_idx, batch in enumerate(batches):
    # Data drift monitoring
    result = stream_monitor.process_batch(batch)

    # Simulate model predictions with confidence
    drift_mag = max(0, (batch_idx - DRIFT_START_BATCH) * 0.12)
    if drift_mag > 0:
        base_conf = 0.85 - min(drift_mag * 0.08, 0.4)
        noise = rng.normal(0, 0.05, BATCH_SIZE)
        simulated_probs = np.clip(
            np.column_stack([1 - (base_conf + noise), base_conf + noise]),
            0, 1,
        )
        simulated_probs = simulated_probs / simulated_probs.sum(axis=1, keepdims=True)
    else:
        simulated_probs = rng.dirichlet([1, 9], size=BATCH_SIZE)

    # Confidence monitoring
    conf_update = confidence_monitor.update(simulated_probs)

    # Correlation tracking
    drift_scores = result.get('scores', {})
    correlation.add_observation(
        confidence=conf_update.get('mean_confidence', 0.5),
        drift_scores=drift_scores,
        entropy=conf_update.get('mean_entropy', 0.5),
        margin=conf_update.get('mean_margin', 0.5),
    )

    all_results.append(result)

print(f"Processed {len(all_results)} batches")
```

## 5. Visualize Drift Progression

Plot the drift scores from all detectors over time.

```python
from production_drift_detection.dashboard.visuals import Visualizer
import plotly.io as pio
pio.renderers.default = 'notebook'

viz = Visualizer()

# Get history
history = stream_monitor.get_history()

# Plot drift scores
fig = viz.plot_drift_scores(history, title='Drift Scores Over Time')
fig.add_vline(x=DRIFT_START_BATCH, line_dash='dot', line_color='orange',
              annotation_text='Drift Started')
fig.show()
```

```python
print("Drift Score Summary:")
print("-" * 50)
for det_name, scores in history.items():
    pre_drift = np.mean(scores[:DRIFT_START_BATCH])
    post_drift = np.mean(scores[DRIFT_START_BATCH:])
    print(f"  {det_name.upper()}: Pre-drift={pre_drift:.4f}, Post-drift={post_drift:.4f}, Ratio={post_drift/max(pre_drift, 1e-10):.2f}x")
```

## 6. Visualize Confidence Degradation

Plot confidence, entropy, and margin over time.

```python
fig = viz.plot_confidence_history(confidence_monitor)
fig.add_vline(x=DRIFT_START_BATCH, line_dash='dot', line_color='orange',
              annotation_text='Drift Started')
fig.show()
```

```python
print("Confidence Summary:")
print("-" * 50)
conf_history = confidence_monitor._confidence_history
entropy_history = confidence_monitor._entropy_history

pre_drift_conf = np.mean(conf_history[:DRIFT_START_BATCH])
post_drift_conf = np.mean(conf_history[DRIFT_START_BATCH:])
pre_drift_ent = np.mean(entropy_history[:DRIFT_START_BATCH])
post_drift_ent = np.mean(entropy_history[DRIFT_START_BATCH:])

print(f"  Confidence: {pre_drift_conf:.4f} -> {post_drift_conf:.4f} ({((post_drift_conf - pre_drift_conf)/pre_drift_conf)*100:.1f}%)")
print(f"  Entropy: {pre_drift_ent:.4f} -> {post_drift_ent:.4f} ({((post_drift_ent - pre_drift_ent)/pre_drift_ent)*100:.1f}%)")
print()
degraded, reason = confidence_monitor.degradation_detected()
print(f"  Degradation detected: {degraded}")
if degraded:
    print(f"  Reason: {reason}")
```

## 7. Confidence-Drift Correlation Analysis

This is the novel research module — analyzing whether confidence
changes precede drift detection.

```python
fig = viz.plot_correlation_analysis(correlation)
fig.show()
```

```python
print("Confidence-Drift Correlation Results:")
print("=" * 50)

corr_summary = correlation.summary()
print(f"Observations: {corr_summary['n_observations']}")
print(f"Early warnings detected: {corr_summary['n_early_warnings']}")
print(f"Early warning score: {corr_summary['early_warning_score']:.2f}/100")
print(f"Confidence leads drift: {corr_summary.get('confidence_is_leading_indicator', 'N/A')}")
print()

# Detailed cross-correlation
for det_name in history:
    cc = correlation.compute_cross_correlation(drift_key=det_name)
    if 'simultaneous_correlation' in cc and cc['simultaneous_correlation']:
        print(f"  {det_name.upper()}: Simultaneous corr={cc['simultaneous_correlation']['correlation']:.4f}, "
              f"Lead={cc.get('confidence_is_leading_indicator', False)}, "
              f"Lag={cc.get('optimal_lag', 'N/A')}")
```

```python
print("Early Warning Score Components:")
print("-" * 50)
warning = correlation.compute_early_warning_score()
for key, value in warning.items():
    if isinstance(value, float):
        print(f"  {key}: {value:.4f}")
    else:
        print(f"  {key}: {value}")
```

## 8. Per-Feature Drift Analysis

Identify which features are most affected by drift.

```python
psi_detector = detectors['psi']
feature_scores = psi_detector.get_feature_scores()

if feature_scores:
    feature_names = psi_detector.feature_names or [f"Feature {i}" for i in range(len(feature_scores[0]))]
    fig = viz.plot_feature_drift_heatmap(feature_scores, feature_names)
    fig.show()
    
    # Latest batch feature ranking
    latest = feature_scores[-1]
    sorted_features = sorted(latest.items(), key=lambda x: x[1], reverse=True)
    print("Most Shifted Features (Latest Batch):")
    print("-" * 40)
    for feature, psi_score in sorted_features:
        severity = "🔴" if psi_score > 0.25 else "🟡" if psi_score > 0.1 else "🟢"
        print(f"  {severity} {feature}: PSI = {psi_score:.4f}")
```

## 9. Detection Latency Analysis

How quickly does each detector identify drift after it's introduced?

```python
print("Detection Latency Analysis:")
print("=" * 50)

for det_name, scores in history.items():
    scores_arr = np.array(scores)
    detector = detectors[det_name]
    latency = compute_detection_latency(
        scores_arr, detector.threshold, DRIFT_START_BATCH
    )
    fpr = compute_false_positive_rate(
        scores_arr, detector.threshold, DRIFT_START_BATCH
    )
    
    print(f"  {det_name.upper()}:")
    print(f"    Threshold: {detector.threshold}")
    print(f"    Drift detected: {latency.get('drift_detected', False)}")
    if latency.get('drift_detected'):
        print(f"    Latency: {latency['detection_latency_batches']} batches")
    print(f"    False positive rate: {fpr['false_positive_rate']:.2%}")
    print()
```

```python
# Plot detection latency
for det_name, scores in history.items():
    scores_arr = np.array(scores)
    detector = detectors[det_name]
    fig = viz.plot_detection_latency(scores_arr, detector.threshold, DRIFT_START_BATCH)
    fig.update_layout(title=f'{det_name.upper()} Detection Latency')
    fig.show()
```

Final evaluation of the monitoring run.

```python
print("=" * 60)
print("  PRODUCTION_DRIFT_DETECTION CASE STUDY — COMPLETE SUMMARY")
print("=" * 60)
print()

print(f"  Total Batches Processed: {N_BATCHES}")
print(f"  Drift Start Batch: {DRIFT_START_BATCH}")
print(f"  Final Drift Magnitude: {drift_magnitude:.2f}")
print()

print("  Drift Detection Summary:")
for det_name, scores in history.items():
    detector = detectors[det_name]
    latency = compute_detection_latency(np.array(scores), detector.threshold, DRIFT_START_BATCH)
    detected = latency.get('drift_detected', False)
    status = '✅ Detected' if detected else '❌ Not detected'
    lat_str = f" (Latency: {latency['detection_latency_batches']} batches)" if detected else ""
    print(f"    {det_name.upper()}: {status}{lat_str}")
print()

print(f"  Confidence Monitor:")
degraded, reason = confidence_monitor.degradation_detected()
print(f"    Degradation detected: {'Yes' if degraded else 'No'}")
if degraded:
    print(f"    Reason: {reason}")
print()

print(f"  Confidence-Drift Correlation:")
print(f"    Early warning score: {correlation.summary()['early_warning_score']:.1f}/100")
print(f"    Early warnings: {len(correlation._early_warnings)}")
print()

print(f"  Total Alerts: {len(alerts)}")
print()
print("=" * 60)
```
